```python
import numpy as np
import matplotlib.pyplot as plt

R = np.zeros((100,100))
G = np.zeros((100,100))
B = np.zeros((100,100))

# Add color to regions
R[:, :50] = 255# left half red
G[25:75, :] = 255 # middle band green
B[50:, 50:] = 255 # bottom-right blue

# Stack into RGB image
img = np.stack([R, G, B], axis=2).astype(np.uint8)

plt.imshow(img)
plt.title("RGB Channel Combination Example")
plt.axis("off")
plt.show()
```
