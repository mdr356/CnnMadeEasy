```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 6x6 grayscale "image"
pixels = np.array([
 [4, 6, 1, 3, 255, 120],
 [0, 9, 7, 3, 2,  90],
 [26, 255, 19, 25, 6,  80],
 [15, 13, 22, 16, 53, 75],
 [1, 8, 4, 3, 10, 60],
 [0, 8, 1, 3, 3,  50]
])

plt.imshow(pixels, cmap='gray')
plt.title("Mini 6×6 Grayscale Image")
plt.colorbar(label='Brightness (0–255)')
plt.show()
```
